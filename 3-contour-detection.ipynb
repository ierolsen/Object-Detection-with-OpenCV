```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/contour.jpg", 0)

plt.figure(figsize=(5,5)), plt.title("Original"), plt.imshow(img, cmap="gray"), plt.axis("off");
```

# Find Contour

```python
contours, hierarch = cv2.findContours(image=img, mode=cv2.RETR_CCOMP, method=cv2.CHAIN_APPROX_SIMPLE)

external_contour = np.zeros(img.shape)
internal_contour = np.zeros(img.shape)

for i in range(len(contours)):
    
    # external
    if hierarch[0][i][3] == -1:
        cv2.drawContours(image=external_contour, contours=contours, contourIdx=i, color=255, thickness = -1)
        
    # internal
    else:
        cv2.drawContours(image=internal_contour, contours=contours, contourIdx=i, color=255, thickness = -1)
        
plt.figure(figsize=(5,5)), plt.title("External Contours"), plt.imshow(external_contour, cmap="gray"), plt.axis("off"); 

plt.figure(figsize=(5,5)), plt.title("Internal Contours"), plt.imshow(internal_contour, cmap="gray"), plt.axis("off"); 
```
