```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/london.jpg",0)

plt.figure(figsize=(5,5)), plt.title("Original"), plt.imshow(img, cmap="gray"), plt.axis("off");
```

# Canny Edge

```python
# simple edge detection
edges = cv2.Canny(image=img, threshold1=0, threshold2=255) #without threshold

plt.figure(figsize=(5,5)), plt.title("Edge"), plt.imshow(edges, cmap="gray"), plt.axis("off");
```

# Determine a Threshold

```python
median_value = np.median(img)
mean_value = np.mean(img)

print(f"median: {median_value} mean: {mean_value}")
```

```python
low = int(max(0,(1 - 0.33)*median_value))

high = int(min(255,(1 + 0.33)*median_value))

print(f"low: {low} high: {high}")
```

```python
edges = cv2.Canny(image=img, threshold1=low, threshold2=high) #with threshold

plt.figure(figsize=(5,5)), plt.title("Edge"), plt.imshow(edges, cmap="gray"), plt.axis("off");
```

# Blurring for All Image

```python
blurred_img = cv2.blur(img, ksize = (6,6))

plt.figure(figsize=(5,5)), plt.title("blurred_img"), plt.imshow(blurred_img, cmap="gray"), plt.axis("off");
```

```python
med_val = np.median(blurred_img)
print(med_val)
```

```python
low = int(max(0, (1 - 0.33)*med_val))

high = int(min(255, (1 + 0.33)*med_val))

print(f"low: {low} high: {high}")
```

```python
edges = cv2.Canny(image = blurred_img, threshold1 = low, threshold2 = high)

plt.figure(figsize=(5,5)), plt.title("blur edge"), plt.imshow(edges, cmap="gray"), plt.axis("off");
```
